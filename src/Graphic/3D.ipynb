```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load data
data = np.loadtxt("option_values.txt")
time = np.unique(data[:, 0])  # Time steps
stock_prices = np.unique(data[:, 1])  # Stock prices

# Reshape data
T, S = np.meshgrid(time, stock_prices)
V = data[:, 2].reshape(len(stock_prices), len(time))

# Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(S, T, V, cmap='viridis')
ax.set_xlabel('Stock Price (S)')
ax.set_ylabel('Time to Maturity (t)')
ax.set_zlabel('Option Price (V)')
plt.title('Option Price Evolution')
plt.show()
```
